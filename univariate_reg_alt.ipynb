Alternatively, the slope of the best fit line can be calculated by

$$m = \frac{\bar{x}.\bar{y} - \bar{xy}}{(\bar{x}^2) - x^2}$$

and the y-intercept of this line is $$b = \bar{y} - m\bar{x}$$

```python
# example from https://www.youtube.com/watch?v=QUyAFokOmow
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')

xs = np.array([1,2,3,4,5,6], dtype=np.float64)
ys = np.array([5,4,6,5,6,7], dtype=np.float64)

plt.scatter(xs, ys)
plt.show()
```

```python
def best_fit_slope_and_intercept(xs, ys):
    m = (mean(xs) * mean(ys) - mean(xs*ys)) / ( mean(xs)*mean(xs) - mean(xs*xs) )
    b = mean(ys) - m * mean(xs)
    return m, b

m, b = best_fit_slope_and_intercept(xs, ys)
```

```python
m, b
```

```python
regression_line = [m*x+b for x in xs]

predict_x = 10
predict_y = (m*predict_x) + b
plt.scatter(xs, ys)
plt.scatter(predict_x, predict_y, color='r')
plt.plot(regression_line)
plt.show()
```
